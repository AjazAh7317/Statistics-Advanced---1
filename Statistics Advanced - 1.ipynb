#1. What is a random variable in probability theory?

->A random variable is a variable whose possible values are outcomes of a random phenomenon. It assigns numerical values to outcomes of a probability experiment.

Example: Tossing a coin → X = {0 if tails, 1 if heads}.

#2. What are the types of random variables?

->Discrete Random Variable – Takes countable values (e.g., number of heads in 10 tosses).

->Continuous Random Variable – Takes uncountable values within an interval (e.g., height, weight, temperature).

#3. Difference between discrete and continuous distributions

->Discrete Distribution → Deals with countable values, probability given by a probability mass function (PMF).

->Continuous Distribution → Deals with uncountable values, probability given by probability density function (PDF).

Example:

Discrete → Binomial distribution.

Continuous → Normal distribution.

#6. What is the Central Limit Theorem (CLT), and why is it critical?

->The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population’s distribution (if sample size ≥ 30).

Importance:

Enables use of normal approximation for inference.

Basis for confidence intervals and hypothesis testing.

#7. What is the significance of confidence intervals in statistical analysis?

->A confidence interval (CI) provides a range of values within which the population parameter is expected to fall with a certain probability (e.g., 95%).

Significance:

Reflects uncertainty in estimates.

More informative than a single point estimate.

Helps in decision making with quantified confidence.

```python
#9.Write a Python program to generate 1000 random numbers from a normaldistribution with mean = 50 and standard deviation = 5. Compute its mean and standarddeviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue')
plt.title("Normal Distribution (Mean=50, Std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#10. Sales Data Analysis using CLT and Confidence Interval
import numpy as np
import scipy.stats as st

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Calculate mean and standard error
mean_sales = np.mean(daily_sales)
std_err = st.sem(daily_sales)

# 95% confidence interval using CLT
ci = st.t.interval(0.95, len(daily_sales)-1, loc=mean_sales, scale=std_err)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", ci)
```
